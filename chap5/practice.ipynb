4장의 실습예제에서 사용했던 광역시도별 인구 데이터 'population.csv'에 대한 분석을 참고하여 다음 물음에 답하시오.
1) 지역별 인구의 히스토그램을 나타내고, 지역별 인구의 평균과 중앙값의 비교 결과에서 얻었던 시사점과 연관지어 설명하시오.
2) 지역별 남녀비율의 히스토그램을 나타내고, 남녀비율의 평균과 중앙값의 비교 결과에서 얻었던 시사점과 연관지어 설명하시오.



```python
#1) 지역별 인구의 히스토그램을 나타내고, 지역별 인구의 평균과 중앙값의 비교 결과에서 얻었던 시사점과 연관지어 설명하시오.
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

data=pd.read_csv('population.csv')
print(data)
pop=data['population']
plt.figure(1)
plt.hist(pop)
plt.figure(2)
plt.boxplot(pop)
plt.show()
print(f'중앙값{np.median(pop)},평균값{np.mean(pop)}')
#histgram에서 le7은 10^7이라는 뜻
#평균값이 중앙값의 약 2배 높음 ,즉 이상치의 존재를 추측해볼 수 있음
```

```python
#2) 지역별 남녀비율의 히스토그램을 나타내고, 남녀비율의 평균과 중앙값의 비교 결과에서 얻었던 시사점과 연관지어 설명하시오.
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

data=pd.read_csv('population.csv')
ratio=[]
for i in range(len(data)):
    male=data.iloc[i]['male']
    femaile=data.iloc[i]['female']
    ratio.append(male/femaile)
plt.hist(ratio)
#다이어그램을 보면 중앙값의 값이 많이 차지하고 있고 양 끝에서는 비슷하게 낮은 빈도를 차지하는 것을 보아 시도별로 남녀비율은 골고로 있는것 같다

```

numpy의 np.random.normal()은 정규분포를 따르는 난수, np.random.exponential()은 지수분포를 따르는 난수를 생성해준다. 이 함수를 이용하여 평균 10, 표준편차
10인 정규분포를 따르는 난수와, 평균 10인 지수분포를 따른 난수를 각각 10,000개씩 발생시켜 이 값들을 상자 그림으로 비교하시오

```python
import numpy as np
import matplotlib.pyplot as plt
x1=np.random.normal(loc =10, scale =10, size = 10000)
x2=np.random.exponential(scale =10, size = 10000)
plt.boxplot([x1,x2], labels=['normal','exponential'])
plt.show()
#np.random.normal(loc =10, scale =10, size = 10000)차례대로 평균,표준편차, 갯수
#np.random.exponential(scale =10, size = 10000) 차례대로 평균, 갯수
#boxplot으로 비교할 때 np로 할때는 [x,y]로 비교하고 pd로 비교할 때는 [[x],[y]]로 비교하는 차이가 있음 
#plt.boxplot([x, y]): 이 방법은 x와 y가 각각 데이터의 모음(예: 리스트, 넘파이 배열, 판다스 시리즈)인 경우 사용됩니다. 
# 입력으로 전달되는 [x, y]는 두 데이터 모음을 포함하는 리스트입니다.

#plt.boxplot([[x],[y]]): 이 방법은 x와 y가 각 데이터 점으로 이루어진 iterable(예: 리스트, 튜플)으로 입력됩니다. 
# 즉, [[x], [y]]는 데이터 점 x와 y를 각각 하나의 데이터 모음으로 만들어 입력합니다.
```

```python
import matplotlib.pyplot as plt

# 임의의 데이터 생성
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# 산점도의 색상과 형태 변경
plt.scatter(x, y, c='red', marker='*')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot with custom color and marker')
plt.show()
#c는 색 marker는 형태
```
