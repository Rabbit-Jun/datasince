자동차 연비와 중량, 실린더 수 등이 나타나 있는 예제 데이터 'auto-mpg.csv'를 대상으로, 가로축에 중량(weight), 세로축에 연비(mpg)를 두고, 실리더 수로 점의 색상을 구분하여 산점도를 그리시오.


```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('auto-mpg.csv')
print(data)
groups=data.groupby('cylinders')

for n,b in groups:
    print(b)
    plt.scatter(b['weight'],b['mpg'],label=n) #b 데이터에서 weight,mpg 열  n=3,4,5,6,7,8 b는 n=3 일 때 데이터 n=4일 때 데이터 ...n=8일 때 데이터

plt.xlabel('Weight')
plt.ylabel('Mpg')
plt.legend()
plt.show()
```

위의 데이터 'auto-mpg.csv'에서 'weight'와 'mpg'를 특성값으로 하여 실린더 수('cylinders')를 예측하는 k=5인 knn 모델을 수립하고, 다음 세 가지 신규 데이터에 대해 실린더 수를 예측하시오.
weight= 2000, mpg= 30
weight= 3500, mpg= 20
weight= 4000, mpg= 15

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('auto-mpg.csv')
print(data)

x_train=data[['weight','mpg']]
y_train=data['cylinders']
from sklearn.neighbors import KNeighborsClassifier
print('<Feature data>')
print(x_train)
print('<Target data>')
print(y_train)
knn =KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
x_pred=pd.DataFrame(data=[[2000,30],[3500,20],[4000,15]] ,columns=['weight','mpg'])
print(x_pred)
print('predict:', knn.predict(x_pred))
```
