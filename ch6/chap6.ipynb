```python
from sklearn import datasets # form <패키지명> import <기능>
dataset =datasets.load_wine() # load_wine()은 네트워크를 통해 wine 데이터를 불러오는 함수
print("feature names=", dataset.feature_names) #와인 데이터 셋에서 feature_names 출력 ,와인 특성의 이름 
print("target=",dataset.target) # 와인 데이터 셋에서 target 출력 , 와인별 등급의 정답 값
print("target names=",dataset.target_names)# 와인 데이터셋에서 target_names 출력 target의 0,1,2각 class_0,class_1,class_2 을 알 수 있다
#특성(feature)과 정답(target)으로 구성된 각각의 와인 데이터를 데이터 포인트(data point)라고 하며 데이터 포인트를 모아놓은 것을 데이터셋(dataset)이라 한다.
```

```python
import pandas as pd
from sklearn import datasets
dataset = datasets.load_wine()
d =pd.DataFrame(dataset.data, columns =dataset.feature_names) 
#pd.DataFrame은 주어진 데이터를 가지고 새로운 데이터프레임을 생성해준다. 
#이 때, columns 옵션에 문자열 리스트를 주면 해당 이름들로 열의 제목을 설정 해준다.
d['target'] =dataset.target #target이라는 새로운 열을 만든 후 dataset.target의 값을 넣어 줌 d의 값의 갯수 = data.target의 갯수가 다르면 오류남
print(d)
```

```python
import pandas as pd
from sklearn import datasets
dataset = datasets.load_wine()
d =pd.DataFrame(dataset.data, columns=dataset.feature_names)
d['target']=dataset.target

import matplotlib.pyplot as plt
plt.scatter(d['alcohol'], d['malic_acid'])
plt.xlabel('alcohol')
plt.ylabel('malic_acid')
plt.show()
```

```python
import pandas as pd
from sklearn import datasets
dataset = datasets.load_wine()
d =pd.DataFrame(dataset.data, columns=dataset.feature_names)
d['target']=dataset.target
groups= d.groupby('target')
for group_name, group_data in groups: #그룹으로 묶은 타겟 데이터를 group_name 그룹 이름 ,group_data 해당 그룹 데이터프레임에 넣음
    print('Group name =', group_name)
    print(group_data)
```

```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt

dataset =datasets.load_wine()
d= pd.DataFrame(dataset.data, columns=dataset.feature_names)
d['target'] =dataset.target
groups = d.groupby('target')
for group_name, group_data in groups: #groups을 이름과 데이터로 나누면 plt에서는 알아서 색상 구분을
    plt.scatter(group_data['alcohol'],group_data['malic_acid'], label=group_name)
plt.legend() #산점도의 우측상단의 범례 표시
plt.show()
```
