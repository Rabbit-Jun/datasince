```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt

dataset =datasets.load_wine()
d= pd.DataFrame(dataset.data, columns=dataset.feature_names)
d['target'] =dataset.target
groups = d.groupby('target')
for group_name, group_data in groups: #groups을 이름과 데이터로 나누면 plt에서는 알아서 색상 구분을
    plt.scatter(group_data['alcohol'],group_data['malic_acid'], label=group_name)
plt.legend() #산점도의 우측상단의 범례 표시
plt.show()
```

```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt

dataset = datasets.load_wine()
d = pd.DataFrame(dataset.data, columns =dataset.feature_names)
d['target']=dataset.target
groups =d.groupby('target')

from mpl_toolkits.mplot3d import Axes3D #matplotlib 패키지 설치할 때 자동으로 함께 설치되는 모듈 3차원 산점도를 그리기 위해 사용한다.  
ax= plt.axes(projection = '3d') #3차원 공간을 투영한 좌표축 설정
for n, d in groups: #name과 data로 넣어준다
    ax.scatter(d['alcohol'], d['malic_acid'], d['hue'], label=n)
plt.legend()
ax.set_xlabel('alcohol')
ax.set_ylabel('malic_acid')
ax.set_zlabel('hue')
plt.show()
# 특성을 보고 정답을 구분하는 모델을 다른 말로 구분 모델이라고도 한다
```

```python
import pandas as pd
from sklearn import datasets
dataset = datasets.load_wine()
d =pd.DataFrame(dataset.data, columns=dataset.feature_names)
d['target'] =dataset.target
x_train = d[['alcohol','malic_acid']] #데이터프레임에서는 여러개읠 열을 추출할 때는 열 제목들을 리스트의 형태로 만들어서 d[] 안에 넣어야 한다 보통 예측에 사용 되는 특성 값을 x로 , 예측에 사용 되는 정답을 y로 표현한다
y_train = d['target']
print("<Feature data>")
print(x_train)
print("<Target data>")
print(y_train)
```

```python
import pandas as pd
from sklearn import datasets
dataset=datasets.load_wine()
d= pd.DataFrame(dataset.data, columns=dataset.feature_names)
d['target']= dataset.target

x_train= d[['alcohol','malic_acid','hue']]
y_train=d['target']
print("<Feature data>")
print(x_train)
print('<Target data>')
print(y_train)


from sklearn.neighbors import KNeighborsClassifier # knn분류 알고리즘 모듈
knn = KNeighborsClassifier(n_neighbors=5) #k를 5로 설정
knn.fit(x_train, y_train) #x는 입력 데이터 y는 입력데이터의 목표 데이터
x_pred=pd.DataFrame(data=[[13,4,13]], columns=['alcohol','malic_acid','hue'])
print(x_pred)
print('Predicted :', knn.predict(x_pred)) #x_pred의 데이터를 가지고 예측
```
